**Categorize Users by Active Minutes**

According to WHO, the physical active minutes for an adult should at least meet one of the following requirements:
* At least 150â€“300 minutes of moderate-intensity aerobic physical activity
* At least 75â€“150 minutes of vigorous-intensity aerobic physical activity
An equivalent combination of moderate- and vigorous-intensity activity throughout the week


```python
--Time spent on activity per day
SELECT t1.id,
       ROUND(AVG(t1.SedentaryMinutes),2) AS avg_SedentaryMinutes,
       ROUND(AVG(t1.VeryActiveMinutes),2) AS avg_VeryActiveMinutes,
       ROUND(AVG(t1.FairlyActiveMinutes),2) AS avg_FairlyActiveMinutes,
       ROUND(AVG(t1.LightlyActiveMinutes),2) AS avg_LightlyActiveMinutes,
       ROUND(AVG(t1.VeryActiveMinutes + t1.FairlyActiveMinutes + t1.LightlyActiveMinutes),2) AS TotalAvg_ActiveMinutes, 
       CASE
           WHEN ROUND(AVG(t1.VeryActiveMinutes + t1.FairlyActiveMinutes + t1.LightlyActiveMinutes),2) BETWEEN 75 AND 150 THEN 'vigorous-intensity'
           WHEN ROUND(AVG(t1.VeryActiveMinutes + t1.FairlyActiveMinutes + t1.LightlyActiveMinutes),2) BETWEEN 150 AND 300 THEN 'moderate-intensity'
       END AS Type_Intensity
FROM extended-cable-393505.bellabeast.daily_activity t1
INNER JOIN extended-cable-393505.bellabeast.sleep_Day t2 ON t1.id = t2.id
--WHERE t2.TotalTimeInBed IS NOT NULL
GROUP BY t1.id;
```
